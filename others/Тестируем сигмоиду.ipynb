```python
import math
import collections as cls
import functools as fn
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))


x = np.linspace(-10, 10, 1000)
y = np.vectorize(sigmoid)(x)
plt.plot(x, y)
```
