```python
import numpy as np
import functools as ft
from matplotlib import pyplot as plt
```

```python
X = np.arange(1, 21)
```

```python
def go(array, function):
    Y = []
    for i in array:
        Y.append(function(i))
    return Y
```

```python
def square(i):
    return i ** 2

print(go(X, square))
plt.scatter(X, go(X, square))
```

```python
def is_prime(n):
    return all(n % i for i in range(2, n))

def simple(n):
    count, number = 0, 2
    last = 0
    while count != n:
        if is_prime(number):
            count += 1
            last = number
        number += 1
    return last
            
print(go(X, simple))
plt.scatter(X, go(X, simple))
```

```python
def pi(i):
    return int(np.pi * 10 ** (i - 1)) / (10 ** (i - 1))
    
print(go(X, pi))
plt.scatter(X, go(X, pi))
```

```python
def is_prime(n):
    return all(n % i for i in range(2, n))

def go_simple(i):
    if is_prime(i):
        return i ** 2
    else:
        return i

print(go(X, go_simple))
plt.scatter(X, go(X, go_simple))
```

```python
def fib(n):
    if n == 1 or n == 2: 
        return 1
    else: 
        return fib(n - 1) + fib(n - 2)
    
print(go(X, fib))
plt.scatter(X, go(X, fib))
```
