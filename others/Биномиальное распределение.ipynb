```python
import numpy as np
from matplotlib import pyplot as plt
from collections import Counter
import random
```

```python
def combo(n, N):
    return np.math.factorial(N) / (np.math.factorial(n) * np.math.factorial(N - n))


def ver(P, n, N):
    return combo(n, N) * P ** n * (1 - P) ** (N - n)


def binom(P, N):
    values = []
    for i in range(N + 1):
        values.append(ver(P, i, N))
    print(values.index(max(values)))
    plt.bar(list(range(N + 1)), values)
```

```python
shari = [1, 2, 3, 4]
counter = []
for i in range(1000000):
    summ = random.choice(shari)
    summ += random.choice(shari[:shari.index(summ)] + shari[shari.index(summ) + 1:])
    counter.append(summ)
    
counter = Counter(counter)
summ = sum(counter.values())

values = []
x = list(counter)

for i in x:
    values.append(counter[i] / summ)
    
plt.bar(x, values, color='#564328')
print(values)
plt.xlabel('Random value')
plt.ylabel('Probability')
```
